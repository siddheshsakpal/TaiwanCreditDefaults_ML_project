```python
# Metrics for each classifier
classifiers = ['XGBClassifier', 'Random Forests', 'Logistic Regression']
accuracy = [0.7865, 0.7865, 0.6862]
precision = [0.5115, 0.5115, 0.3734]
recall = [0.6059, 0.6059, 0.6340]
f1_score = [0.5547, 0.5547, 0.4700]
auc_score = [0.7216, 0.7216, 0.6674]

# Plotting the metrics
barWidth = 0.2
r1 = np.arange(len(classifiers))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]

plt.figure(figsize=(12, 6))

plt.bar(r1, accuracy, color='b', width=barWidth, edgecolor='grey', label='Accuracy')
plt.bar(r2, precision, color='g', width=barWidth, edgecolor='grey', label='Precision')
plt.bar(r3, recall, color='r', width=barWidth, edgecolor='grey', label='Recall')
plt.bar(r4, f1_score, color='c', width=barWidth, edgecolor='grey', label='F1-Score')
plt.bar(r5, auc_score, color='m', width=barWidth, edgecolor='grey', label='AUC-Score')

# Adding labels
plt.xlabel('Classifiers', fontweight='bold', fontsize=15)
plt.xticks([r + barWidth*2 for r in range(len(classifiers))], classifiers)

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
